```python
class Patient:
    last_patient_id = 0
    patients_list = []
    def __init__(self,name,age,gender,med_history,contact_info, address):
        Patient.last_patient_id += 1
        self.patient_id=Patient.last_patient_id
        self.name=name
        self.age=age
        self.gender=gender
        self.med_history=med_history
        self.contact_info=contact_info
        self.address=address
        Patient.patients_list.append(self)
    def update_data(self,name=None, age=None, gender=None, med_history=None):
        if name is not None:
            self.name=name
        if age is not None: 
            self.age=age
        if gender is not None:
            self.gender=gender
        if med_history is not None:
            self.med_history=med_history
        print("Successed!")
    
    def patient_info(self):
        print(f'Patient info: \nName: {self.name}\nAge: {self.age}\nGender: {self.gender}\nmed_history: {self.med_history}.')
    @classmethod
    #cls==> class
    def display_patients(cls):
        if len(cls.patients_list)==0:
            print("Unavailable.")
        else:
            print("List of Patients: ")
            for P in cls.patients_list:
                print(f'Patient ID: {P.patient_id}\t Name: {P.name}.')
    @classmethod
    def patient_by_name(cls,name):
        for p in cls.patients_list:
            if p.name==name:
                 return p
        print("Patient not found.")
        return None
```

```python
#Patient Class Objects
patient1 = Patient(name="Edward", age=30, gender="Male", med_history="Hypertension", contact_info="01001234567", address="Banha")
patient2 = Patient(name="Lily", age=25, gender="Female", med_history="Diabetes", contact_info="01007654321", address="Giza")
print("===")

#Display a specific patient info
patient1.patient_info()
print("===")

#Display all information
Patient.display_patients()
print("===")

#Update patient data
patient1.update_data(age=32, med_history="Hypertension, Asthma")
```

```python
class Doctor:
    last_doctor_id=0
    doctors_list=[]
    def __init__(self,name,specialization):
        Doctor.last_doctor_id+=1
        self.doctor_id=Doctor.last_doctor_id
        self.name=name
        self.specialization=specialization
        self.schedule = [] 
        Doctor.doctors_list.append(self)
        
    @classmethod
    def doctor_by_id(cls,doctor_id):
        for d in cls.doctors_list:
            if doctor_id==d.doctor_id:
                return d
        return None
    
    @classmethod
    def display_doctors(cls):
        if len(cls.doctors_list)==0:
            print("Unavailable.")
        else:
            print("List of doctors: ")
            for d in cls.doctors_list:
                print(f'Doctor ID: {d.doctor_id}\t Specialization: {d.specialization}.')
            
    @classmethod
    def display_information_by_name(cls,name):
        for d in cls.doctors_list:
            if d.name==name:
                print(f"Dr. {name}:\nID:{d.doctor_id}\t Specialization: {d.specialization}.")
            
    def add_to_schedule(self,doctor_id,date,time,status):
        doc= Doctor.doctor_by_id(doctor_id)
        if doc:
            doc.schedule.append({
                "Doctor ID": doctor_id,
                "Date": date,
                "Time": time,
                "Status": status
            })
            print(f"Added to Dr. {doc.name}'s Schedule.")
        else:
            print("Doctor not found.")
            
    def display_doctor_schedule(self,doctor_id):
        doc= Doctor.doctor_by_id(doctor_id)
        if doc:
            if doc.schedule:
                print(f"Dr. {doc.name}'s schedule: ")
                for a in doc.schedule:
                    print(f"Date: {a['Date']}\t Time: {a['Time']}\t Status: {a['Status']}.")
            else:
                print(f"No schedule available for Dr. {doc.name}.")
        else:
            print("Doctor not available.")
            
    def modify_schedule(self,doctor_id,date,time, n_status=None, n_patient=None):
        doc=Doctor.doctor_by_id(doctor_id)
        if doc:
            for p in doc.schedule:
                if p['Date']==date and p['Time']==time:
                    if n_status:
                        p['Status']=n_status
                    print("Schedule updated successfully!")
                    return
            print("Not found.")
        print("No doctor available.")
        
    def remove_from_schedule(self,doctor_id,date,time):
        doc=Doctor.doctor_by_id(doctor_id)
        if doc:
            for p in doc.schedule:
                if p['Date']==date and p['Time']==time:
                    doc.schedule.remove(p)
                    print(f"Appointment on {date} at {time} has been removed for Dr. {doc.name}.")
                    return
            print(f"No appointment has been found on {date} at {time}.")
        else:
            print("Not found.")
```

```python
#Doctor class Objects
doctor1 = Doctor(name="Menna", specialization="Psychiatrist")
doctor2 = Doctor(name="James", specialization="Neurology")
print("===")

#Display all information
Doctor.display_doctors()
print("===")

# Add appointments to the doctor's schedule
doctor1.add_to_schedule(doctor_id=1, date="2025-02-01", time="10:00 AM", status="Available")
doctor2.add_to_schedule(doctor_id=2, date="2025-02-02", time="02:00 PM", status="Available")
print("===")

#Display a specific doctor info 
doctor1.display_doctor_schedule(doctor_id=1)
print("===")

#Modify schedule for a doctor
doctor1.modify_schedule(doctor_id=1, date="2025-02-01", time="10:00 AM", n_status="Booked")
print("===")

#Remove an appointment from the schedule
doctor1.remove_from_schedule(doctor_id=1, date="2025-02-01", time="10:00 AM")
```

```python
class Appointment:
    last_appointment_id=0
    appointment_list=[]
    def __init__(self,date,time,status='available'):
        Appointment.last_appointment_id+=1
        self.appointment_id=Appointment.last_appointment_id
        self.date=date
        self.time=time
        self.status=status
        self.patient = None 
        self.doctor = None  

    @classmethod
    def add_appointment(cls,appointment_id,patient_id,doctor_id,date,time):
        #any() function returns True if any item in an iterable are true, otherwise it returns False.
        #The next() function returns the first matching item from an iterator, or a default value (like None) if no match is found.
        patient = next((p for p in Patient.patients_list if p.patient_id == patient_id), None)
        doctor = next((d for d in Doctor.doctors_list if d.doctor_id == doctor_id), None)

        if not patient:
            print(f"Patient that has the ID {patient_id} doesn't exist.")
            return
        
        if not doctor:
            print(f"Doctor that has the ID {doctor_id} doesn't exist.")
            return
            
        for p in cls.appointment_list:
            if p.date==date and p.time==time:
                if p.status.lower()=='available':
                    p.status='Booked'
                    p.patient = patient 
                    p.doctor = doctor
                    print(f"Appointment on {date} at {time} is now booked for Patient ID {patient_id} with Doctor ID {doctor_id}.")
                    return
                else:
                    print(f'Appointment on {date} is already booked.')
                    return
                
        #if there's no appointment with the same data then add new one
        new_appointment = Appointment(date, time, "Booked")
        new_appointment.patient = patient
        new_appointment.doctor = doctor    
        cls.appointment_list.append(new_appointment)
        print(f"New appointment has been added successfully for Patient ID {patient_id} with Doctor ID {doctor_id} on {date} at {time}.")
                
    @classmethod
    def appointments_for_doctors(cls,doctor_id):
        appointments=[a for a in cls.appointment_list if a.doctor.doctor_id == doctor_id]
        if appointments:
            for a in appointments:
                print(f"Appointment on {a.date} at {a.time} for {a.patient.patient_id}, status: {a.status}.")
        else:
            print(f"No appointments for this doctor available.")
```

```python
#Appointment class Objects
Appointment.add_appointment(appointment_id=1, patient_id=1, doctor_id=1, date="2025-02-01", time="10:00 AM")
Appointment.add_appointment(appointment_id=2, patient_id=2, doctor_id=1, date="2025-02-01", time="02:00 PM")
print("===")

#Display a specific doctor's schedule
Appointment.appointments_for_doctors(doctor_id=1)
print("===")

#Attempting to add appointment at the same date
Appointment.add_appointment(appointment_id=3, patient_id=3, doctor_id=1, date="2025-02-01", time="10:00 AM")
```

```python
class Billing:
    last_bill_id=0
    bill_list=[]
    def __init__(self,patient_id,appointment_id,total_amount=0,payment_state='None'):
        Billing.last_bill_id+=1
        self.bill_id=Billing.last_bill_id
        self.patient_id=patient_id
        self.appointment_id=appointment_id
        self.total_amount=total_amount
        self.payment_state=payment_state
        Billing.bill_list.append(self)
        
    def calculate_total_amount(self, services):
        prices={
            "consultation": 100,
            "x-ray": 500,
            "blood_test": 250,
            "medication": 30
        }
        self.total_amount=sum(prices[n] for n in services)
        print(f"Total amount: {self.total_amount}")
    
    def mark_as_paid(self):
        self.payment_state='Paid'
        
    def display_a_bill(self):
        print(f"Bill ID: {self.bill_id}\tPatient ID: {self.patient_id}")
        print(f"Appointment ID: {self.appointment_id}\tTotal Amount: {self.total_amount}")
        print(f"Payment State: {self.payment_state}")
     
    @classmethod   
    def find_bill(cls,bill_id):
        for b in cls.bill_list:
            if b.bill_id==bill_id:
                return b
        print(f"Bill ID: {bill_id} Not found.")
        return None
    
    @classmethod
    def total_amount_overall(cls):
        total = sum(b.total_amount for b in cls.bill_list if b.payment_state == 'Paid')
        print(f"Total amount from all paid bills: {total}")
        
    @classmethod
    def delete_bill(cls,bill_id):
        b=cls.find_bill(bill_id)
        if b:
            cls.bill_list.remove(b)
            print(f"Bill ID: {bill_id} has been deleted successfully.")
            return
        else:
            print(f"Bill ID: {bill_id} not found.")
    
    @classmethod
    def display_all_bills(cls):
        for b in cls.bill_list:
            b.display_a_bill()
            print("==")

            
```

```python
#Billing Class Objects
bill1 = Billing(patient_id=1, appointment_id=1, total_amount=0, payment_state='None')
bill2 = Billing(patient_id=2, appointment_id=2, total_amount=0, payment_state='None')
print("===")

#Calculating the total amount for services
bill1.calculate_total_amount(["consultation", "x-ray", "blood_test"])
bill2.calculate_total_amount(["consultation", "medication"])
print("===")

#Marking the bill as paid
bill1.mark_as_paid()
bill2.mark_as_paid()
print("===")

#Displaying the bill details
bill1.display_a_bill()
bill2.display_a_bill()
print("===")

#Displaying all bills
Billing.display_all_bills()
print("===")

#Displaying the total amount from all paid bills
Billing.total_amount_overall()
print("===")

#Deleting a bill
Billing.delete_bill(bill_id=1)

#Displaying all bills after deletion
Billing.display_all_bills()
```
